```python
# –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫\nimport pandas as pd\nimport matplotlib.pyplot as plt\nimport seaborn as sns\nimport warnings\nwarnings.filterwarnings('ignore')\n\n# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è\nsns.set(style=\"whitegrid\")\nplt.rcParams['figure.figsize'] = (12, 6)\n\nprint(\"‚úÖ –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã\")
```

```python
# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö\nsocial = pd.read_csv(\"C:/SQL/social.csv\", delimiter=';')\neconomic = pd.read_csv(\"C:/SQL/economic.csv\", delimiter=';')\n\nprint(\"‚úÖ –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã\")\nprint(f\"Social: {social.shape[0]} —Å—Ç—Ä–æ–∫, {social.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤\")\nprint(f\"Economic: {economic.shape[0]} —Å—Ç—Ä–æ–∫, {economic.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤\")
```

## 1. –ê–Ω–∞–ª–∏–∑ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ —Ç–æ–ø–æ–≤\ndef print_top_bottom(df, column, n=3):\n    print(f\"\\nüìä {column}:\")\n    top = df.nlargest(n, column)[['municipal', column]]\n    bottom = df.nsmallest(n, column)[['municipal', column]]\n    print(f\"üî∫ –õ—É—á—à–∏–µ {n}:\\n{top.to_string(index=False)}\")\n    print(f\"üîª –•—É–¥—à–∏–µ {n}:\\n{bottom.to_string(index=False)}\")\n    return top, bottom\n\n# –°–æ—Ü–∏–∞–ª—å–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏\nsocial_metrics = ['zp', 'kolichestvo_zhilyo', 'kolichestvo_lpu']\nsocial_tops = {}\n\nfor metric in social_metrics:\n    top, bottom = print_top_bottom(social, metric)\n    social_tops[metric] = {'top': top, 'bottom': bottom}
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π\nfor metric in social_metrics:\n    plt.figure(figsize=(10, 6))\n    df_sorted = social.nlargest(20, metric)\n    sns.barplot(x=metric, y='municipal', data=df_sorted, palette='viridis')\n    plt.title(f'–¢–æ–ø-20 —Ä–∞–π–æ–Ω–æ–≤ –ø–æ {metric}')\n    plt.xlabel(metric)\n    plt.ylabel('–ú—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω–æ–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ')\n    plt.tight_layout()\n    plt.show()
```

## 2. –ê–Ω–∞–ª–∏–∑ —ç–∫–æ–Ω–æ–º–∏—á–µ—Å–∫–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π

```python
# –≠–∫–æ–Ω–æ–º–∏—á–µ—Å–∫–∏–µ –º–µ—Ç—Ä–∏–∫–∏\neconomic_metrics = ['kol_pred', 'pribyl', 'invest']\neconomic_tops = {}\n\nfor metric in economic_metrics:\n    top, bottom = print_top_bottom(economic, metric)\n    economic_tops[metric] = {'top': top, 'bottom': bottom}
```

```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —ç–∫–æ–Ω–æ–º–∏—á–µ—Å–∫–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π\nfor metric in economic_metrics:\n    plt.figure(figsize=(10, 6))\n    df_sorted = economic.nlargest(20, metric)\n    sns.barplot(x=metric, y='municipal', data=df_sorted, palette='coolwarm')\n    plt.title(f'–¢–æ–ø-20 —Ä–∞–π–æ–Ω–æ–≤ –ø–æ {metric}')\n    plt.xlabel(metric)\n    plt.ylabel('–ú—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω–æ–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ')\n    plt.tight_layout()\n    plt.show()
```

## 3. –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–π

```python
# –¢–æ–ø-10 –ø–æ –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏—è–º + –æ—Å—Ç–∞–ª—å–Ω—ã–µ\ninvest_sorted = economic.nlargest(10, 'invest')[['municipal', 'invest']]\nother_value = economic['invest'].nsmallest(len(economic)-10).sum()\nother_row = pd.DataFrame({'municipal': ['–û—Å—Ç–∞–ª—å–Ω—ã–µ'], 'invest': [other_value]})\ninvest_final = pd.concat([invest_sorted, other_row])\n\n# –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞\nplt.figure(figsize=(10, 8))\ncolors = sns.color_palette('Set3', len(invest_final))\nplt.pie(invest_final['invest'], \n        labels=invest_final['municipal'], \n        autopct='%1.1f%%', \n        startangle=140,\n        colors=colors)\nplt.title('–î–æ–ª—è –∏–Ω–≤–µ—Å—Ç–∏—Ü–∏–π –≤ –æ—Å–Ω–æ–≤–Ω–æ–π –∫–∞–ø–∏—Ç–∞–ª –ø–æ –º—É–Ω–∏—Ü–∏–ø–∞–ª—å–Ω—ã–º –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è–º (–¢–æ–ø-10 + –æ—Å—Ç–∞–ª—å–Ω—ã–µ)')\nplt.axis('equal')\nplt.tight_layout()\nplt.show()
```
